```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('demand_encoded.csv')

X = df.drop(columns=['demand'])
y = df['demand']

model = LinearRegression()
model.fit(X, y)

print("Model trained on full dataset.")
print(f"Features: {X.columns.tolist()}")
```

```python
def predict_demand(mineral, technology, year):
    # start with a row of all zeros
    row = pd.DataFrame(columns=X.columns, data=[[0]*len(X.columns)])
    
    # set year
    row['year'] = year
    
    # set mineral flag
    mineral_col = f'mineral_{mineral}'
    if mineral_col in row.columns:
        row[mineral_col] = 1
    else:
        print(f"Unknown mineral: {mineral}")
        return None
    
    # set technology flag
    tech_col = f'technology_{technology}'
    if tech_col in row.columns:
        row[tech_col] = 1
    else:
        print(f"Unknown technology: {technology}")
        return None
    
    prediction = model.predict(row)[0]
    return round(prediction, 1)
```

```python
print(predict_demand('Copper', 'Solar PV', 2030))
print(predict_demand('Lithium', 'Electric vehicles', 2035))
print(predict_demand('Nickel', 'Wind', 2040))
print(predict_demand('Cobalt', 'Electric vehicles', 2024))
```

```python
print(predict_demand('Copper', 'Electric vehicles', 2045))
print(predict_demand('Lithium', 'Electric vehicles', 2050))
```
